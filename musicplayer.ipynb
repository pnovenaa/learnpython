```python
from pygame import mixer
from tkinter import *

root = Tk()
root.geometry("600x300")

mixer.init()
mixer.music.load("cardigan_whatitsworth.MP3")


def pause():
    mixer.music.pause()


def play():
    mixer.music.play()


def resume():
    mixer.music.unpause()


Label(root, text=''' /ᐠ｡ꞈ｡ᐟ\ @pnovenaa's music player ✧/ᐠ-ꞈ-ᐟ\ ''', font="lucidia 15 bold").pack()
Button(text="Play", command=play,bg="#ff9999").place(x=200, y=100)
Button(text="Pause", command=pause,bg="#ff9999").place(x=250, y=100)
Button(text="Resume", command=resume,bg="#ff9999").place(x=310, y=100)
Button(text="Quit", command=quit,bg="#ff9999").place(x=380, y=100)

root.mainloop()
```
